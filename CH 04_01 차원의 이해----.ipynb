## 2. 차원과 Dataset

- 차원(dimension)은 인공지능과 data science의 핵심 주제로,  data에 포함된 feature들의 수임<br>
\- 예를 들어 어린 아이의 키에 영향을 미치는 변수 $z$가 $z =\left( 부모의 \ 유전자, 1일 \ 영양섭취량 \right)$로 가정할 때 <br> $ \ \ $  feature  $z$는 2차원 data임<br><br>

- $n_{_0}$개 feature $z = \left( z_1, z_2, ..., z_{n_{_0}} \right)$에 대해 $n$번의 관측을 한 경우, $i^{th}$ 관측치 $z_i = \left( z_{i,1}, z_{i,2}, ..., z_{i,n_{_0}} \right)$와 <br> $ \ \ $ label $y_i$가 얻어지게 됨

- 관측 dataset $\left( Z, y \right)$를 행렬로 표현하면 다음과 같음

$$
Z_{n \times n_{_0}}=
\begin{bmatrix}
    z_{11} & z_{12} & \cdots & z_{1n_{_0}} \\
    z_{21} & z_{22} & \cdots & z_{2n_{_0}} \\
    \vdots & \vdots & \ddots & \vdots \\
    z_{n1} & z_{n2} & \cdots & z_{nn_{_0}}
\end{bmatrix}
, \qquad
y_{n \times 1}=
\begin{bmatrix}
    y_{1} \\
    y_{2} \\
    \vdots \\
    y_{n} 
\end{bmatrix}
$$

### 3.2 거리함수

- 특정 공간(차원)에서 점들간의 거리에 따라 같은 class에 속하거나 다른 class에 속한다고  생각할 수 있음<br>
\- 같은 class의 data끼리는 거리가 짧고 다른 class의 data간에는 거리가 멀다고 생각하는 것임 

이름 | $\qquad \qquad \qquad$ 공식 |  $\qquad$ 부가설명
:----:|:----|:----
<br> $p-$Norm | <br> $$
L_p = \left( \sum_{i=1}^{n} \left| z_{1i} - z_{2i} \right|^p \right)^{\frac{1}{p}} $$ | <br> $\bullet \ $  Manhattan distance: $p=1$ <br> $\bullet \ $ Euclidean distance: $p=2$ <br> $\bullet \ $Chebyshev distance:  $p \ \rightarrow \ \infty$
<br> Cosine <br> distance | <br> $$
cos(\theta)  = \frac{p_1 \cdot p_2}{ \| p_1 \| \| p_2 \|} $$ | <br> $\bullet \ $ $ \cdot $은 내적(inner product임) <br> $\bullet \ $ vector의 크기 보다는 각도에 따라 거리를 측정하는 개념임 <br> $\bullet \ $ 문서(document, text)분석에 가장 많이 사용되는 거리
<br> Pearson  <br> correlation  <br> distance | <br> $$ \rho_{_{z_1, z_2}} = \frac{cov\left( z_1, z_2 \right)}{\sigma_{z_1}\sigma_{z_2}}
$$ | <br> $\bullet \ $ 2개의 확률변수간에 (선형)상관관계를 기반으로한 거리측정
<br> Mahalanobis  <br> distance | <br> $$D_M = exp \left( - \frac{1}{2}(z-\mu)'\Sigma^{-1} (z-\mu) \right)$$ | <br> $\bullet \ $ $\Sigma$는 covariance matrix  <br> $\bullet \ $ 높은 계산 복잡도
<br> Frobenius  <br> Norm | <br> $$
\begin{align}
\| A \|_F =& \sqrt{\sum_{i=1}^{m}\sum_{i=1}^{n}a_{ij}^2} \\
          =& \sqrt{ tr \left( A^{T}A \right)}
\end{align}
$$ | <br> $\bullet \ $ $m \times n$ 행렬 $A$에 대한 거리
<br> Kullback  <br> Leibler  <br> Divergence | <br> $$ 
\begin{align}
D_{KL}(P \| Q) =& 
\begin{cases} 
\sum_i log_2 \left( \frac{P(i)}{Q(i)} \right) P(i)     &       if \ 이산형확률변수  \\\\
\int_i log_2 \left( \frac{p(x)}{q(x)} \right) p(x) dx   &       if \ 연속형확률변수
\end{cases}
\end{align}
$$ | <br>  $\bullet \ $  확률분포함수 $P$와  $Q$의 차이를 측정하는 measure  <br>  $\bullet \ $ 단,  $P$와  $Q$의 위치에 따라 값이 다름에 주의 $$ D_{KL}(P \| Q) \ne D_{KL}(Q \| P)$$
<br> Cross  <br> Entropy | <br> $$ 
\begin{align}
H_p(q) =& E_p \left( log \frac{1}{q} \right) \\
       =&
\begin{cases} 
\sum_i log \frac{1}{q(i)} p(i)     &       if \ 이산형확률변수  \\\\
\int_x log \frac{1}{q(y)} p(y)dy   &       if \ 연속형확률변수
\end{cases}
\end{align}
$$  | <br>  $\bullet \ $  Kullback Leibler  Divergence $D_{KL}(P \| Q)$에 상수를 <br> $ \ \ $ 더한 값으로 $D_{KL}(P \| Q)$와 같은 의미의 measure임

### 3.4  거리함수(Distance Measure)

#### (1) $Norm$

#### (1.1) $Norm$의 정의

- 임의의 scalar $a$와 임의의 $n-$차원 벡터(또는 point) $p_1, p_2$에 대해서$ \left(p_1, p_2 \in R^n \right)$, <br> $ \ \ $ 다음 4가지 조건을 만족하는 함수 $f: R^n \rightarrow R$를 norm이라고 함

>$
\begin{align}
1. \ & Non-negativity  \qquad \ \ \   \forall p \in R^n, f \left( p \right) \ge 0  \\
2. \ & Homogeneity \qquad \qquad \   f \left( ap \right) = |a| f\left( p\right) \\
3. \ & Positive \ definiteness \quad \ if \ \ f(p) = 0 이면 \ \ p=0 \\
4. \ & Triangular \ Inequality \quad f\left( p_1 + p_2 \right) \le f\left( p_1 \right) + f\left( p_2 \right) \\
\end{align}
$

#### (1.2) $p-$norm 

- $n-$차원인 두 점 $z_1= \left( z_{11}, z_{12}, ..., z_{1n} \right)$과  $z_2= \left( z_{21}, z_{22}, ..., z_{2n} \right)$ 사이의 $L_p$는 다음과 같이 정의 됨

$$
L_p \left( z_1, z_2 \right) = \left( \sum_{i=1}^{n} \left| z_{1i} - z_{2i} \right|^p \right)^{\frac{1}{p}}
$$


```python
import numpy as np

point1 = np.array((0, 1))
point2 = np.array((1, 0))

  

# n: vector 크기
# p1, p2 vector
# p: p-norm

def Lp(p1, p2, p):
    
    
    if np.array_equal(p1, p2):       # 만일 두 벡터가 같으면
        sum = 0
        for i in range(len(point1)):
            sum = sum + (np.abs( p1[i]) )**p
            
    else:
        sum = 0
        for i in range(len(point1)):
            sum = sum + (np.abs( p1[i] - p2[i] ))**p
        
    return sum**(1/p)
```

```python
Lp(point1, point2, 1)
```

```python
Lp(point1, point2, 2)
```

```python
Lp(point1, point1, 2)
```

#### (1.3)  Manhattan Distance 

- $n-$차원인 두 점 $z_1= \left( z_{11}, z_{12}, ..., z_{1n} \right)$과  $z_2= \left( z_{21}, z_{22}, ..., z_{2n} \right)$ 사이의 Manhattan Distance$(L_1)$은 다음과 같이 정의 됨

$$
L_1 \left( z_1, z_2 \right) = \sum_{i=1}^{n} \| z_{1i} - z_{2i} \|
$$

- 수직 이동거리와 수평 이동거리의 합임

$\quad$ $\Longrightarrow$ 즉 $L_1-$ norm은 $L_p-$ norm에서 $p=1$인 특별한 경우임

```python
eps = 1.0e-8

def L1(p1, p2): 
    
    if np.array_equal(p1, p2):       # 만일 두 벡터가 같으면
        sum = 0
        for i in range(len(point1)):
            sum = sum + (np.abs( p1[i]) )**p
    else:
        sum = 0
        for i in range(len(p1)):
            sum = sum + np.abs(p1[i]-p2[i])
    
    return sum
```

```python
print(L1(point1, point2))
```

```python
np.abs(Lp( point1, point2, 1) - L1(point1, point2)) < eps
```

#### (1.4) Euclidean Distance

- $n-$차원인 두 점 $p_1= \left( p_{11}, p_{12}, ..., p_{1n} \right)$과  $p_2= \left( p_{21}, p_{22}, ..., p_{2n} \right)$ 사이의 Euclidean distence $(L_2)$는 다음과 같이 정의 됨

$$
L_2 \left( p_1, p_2 \right) = \sqrt{\sum_{i=1}^{n} \left( p_{1i} - p_{2i} \right)^2}
$$

$\quad$ $\Longrightarrow$ 즉 $L_2-$ norm은 $L_p-$ norm에서 $p=2$인 특별한 경우임

```python
import numpy as np

point1 = np.array((0, 1))
point2 = np.array((1, 0))
```

```python
def L2(p1, p2):
    if np.array_equal(p1, p2):       # 만일 두 벡터가 같으면
        sum = 0
        for i in range(len(point1)):
            sum = sum + (np.abs( p1[i]) )**2
            
    else:
        sum = 0
        for i in range(len(point1)):
            sum = sum + (np.abs( p1[i] - p2[i] ))**2
        
    return sum**(1/2)
```

```python
L2(point1, point2)
```

```python
eps = 1.0e-8
np.abs(Lp(point1, point2, 2) - L2(point1, point2)) < eps
```

```python
Lp(point1, point2, 2), L2(point1, point2)
```

#### (1.5) Mahalanobis distance $(D_M)$

- 하나의 확률변수 $Z$에 대한 정규분포 함수

$$
f(z) = \frac{ 1 }{\sqrt{2 \pi } \sigma } exp \left( - \frac{(z-\mu)^2}{2\sigma^2} \right)
$$

- 여러개의 확률변수 $ Z= \left( z_1, z_2, ..., z_n \right)$의 확률분포(multivariate normal distribution)함수

$$
f(z) = \frac{ 1 }{\sqrt{(2 \pi)^n | \Sigma | }} exp \left( - \frac{1}{2}(z-\mu)'\Sigma^{-1} (z-\mu) \right)
$$

$\qquad$ 여기서 $\mu = E(Z)$이고 $\Sigma = Cov(Z, Z)$

- Mahalanobis distance $D_M$

$$
D_M = exp \left( - \frac{1}{2}(z-\mu)'\Sigma^{-1} (z-\mu) \right)
$$

#### (2)  코사인 거리(Cosine Distance) 

#### (2.1) 내적

- $n-$차원의 두 벡터 $z_1$과 $z_2$의 코사인 유사도(Cosine Distance )는 내적공간의 두 벡터간 각도 $\theta$와 $cos(\theta)$값을 이용하여 측정된 vector간 각도의 차이임
- 코사인 유사도는 '방향'에 대한 거리임

$$
\begin{align}
D_{cos} \left( z_1, z_2 \right) =& cos(\theta) \\

#### (3) Frobenius Norm

- Matrix를 대상으로한 norm으로 다음과 같이 정의 됨
- $m \times n$ matrix $ A = \left( a_{ij} \right)$에 대해 

$$
\begin{align}
\| A \|_F =& \sqrt{\sum_{i=1}^{m}\sum_{i=1}^{n}a_{ij}^2} \\
          =& \sqrt{ tr \left( A^{T}A \right)}
\end{align}
$$

```python
def F_norm(m, n, M):
    sum = 0
    for i in range(m):
        for j in range(n):
            sum = sum + M[i, j]**2 
    return np.sqrt(sum)
```

다음과 같이 결합확률분포의 Entropy로 확장이 가능

$$
\begin{align}
H(X,Y) =& E[I(X, Y)] \\ 
       =& \sum_{y} \sum_{x}  \log_2 \left( \frac{1}{P(x,y)} \right) P(X=x,Y=y)
\end{align}
$$

만일 $X$와 $Y$가 통계적 독립이라면 Entropy는 다음과 같음

$$
\begin{align}
H(X,Y) =& E[I(X, Y)] \\ 
       =& \sum_{y} \sum_{x} log_2 \left( \frac{1}{P(x, y)} \right) P(X=x, Y=y) \\
       =& \sum_{y} \sum_{x} log_2 \left( \frac{1}{P(x)P(y)} \right) P(X=x)P(Y=y) \\
       =& \sum_{y} \sum_{x} \left(log_2 \left( \frac{1}{P(x)} \right) + log_2 \left( \frac{1}{P(y)}                   \right) \right)P(X=x)P(Y=y) \\
       =& \sum_{y} \sum_{x} log_2 \left( \frac{1}{P(x)} \right) P(X=x)P(Y=y)
           +
          \sum_{y} \sum_{x}  log_2 \left( \frac{1}{P(y)} \right) P(X=x)P(Y=y) \\
       =& \sum_{y} P(Y=y) \sum_{x} log_2 \left( \frac{1}{P(x)} \right) P(X=x)
           +
          \sum_{y} P(X=x) \sum_{x}  log_2 \left( \frac{1}{P(y)} \right) P(Y=y) \\
       =& \sum_{y} P(Y=y) H(X) + \sum_{y} P(X=x) H(Y) \\
       =& H(X)+H(Y)
\end{align} 
$$

즉, 2개의 확률변수가 통계적 독립이면 결합 Entropy는 각각의 Entropy의 합과 같음

#### (6)  Kullback-Leibler Divergence

#### (6.1) Kullback-Leibler Divergence의 정의

- Kullback-Leibler Divergence는 기계학습에서 2개의 확률분포 간의 차이를 계산하는 매우 중요한 정의임<br><br>

- 쿨백-라이블러 발산(Kullback–Leibler divergence, KLD)은 <span style="color:red">두 확률분포의 차이</span>를 계산하는 함수로, <br> $ \ \ $  <span style="color:blue">이상적인 분포 $P$에 대해, 그 분포를 근사하는 다른 분포 $Q$를 사용해 샘플링을 한다면 발생할 수 있는 <br> $ \ \ $ information entropy 차이를 계산</span><br>
$\quad$ $\checkmark \ $ 기계학습에서 이상적인 분포 $P$는 label의 분포를 의미하고 $Q$는 feature를 이용하여 label의 추정한  $\hat y$의 <br> $\qquad$ 분포를 의미함<br>
$\quad$ $\checkmark \ $ 분포 $P$와 분포 $Q$는의 차이가 작다면  feature를 이용하여 label $y$를 잘 근사화(approximation)했다고 할 수 있음<br>
$\quad$ $\checkmark \ $ 상대 엔트로피(relative entropy), 정보 획득량(information gain), 인포메이션 다이버전스(information  <br> $\qquad$ divergence)라고도 함<br>
$\quad$ $\checkmark \ $ 정보이론에서는 상대 엔트로피, 기계학습의 결정 트리에서는 정보 획득량을 주로 사용함<br><br>

- 쿨백-라이블러 발산은 비대칭으로, 두 값의 위치를 바꾸면 함수값도 달라진다. 따라서 이 함수는 거리 함수는 아님

$$
D_{KL}(P \| Q) \neq D_{KL}(Q \| P)
$$

- 동일한 집합에 대해 정의된 두 확률 분포 $P$ 및 $Q$ 간의 Kullback-Leibler divergence는 다음과 같이 정의됨

**[정의] Kullback-Leibler Divergence**<br><br>

>두 확률변수에 대한 분포 $P$와  $Q$가 있을 때, 두 분포의 Kullback-Leibler Divergence은 다음과 같이 정의된다<br>
>
>$\quad$ $\bullet$ Discrete Random Variable인 경우
$$
D_{KL}(P \| Q) = \sum_i log_2 \left( \frac{P(i)}{Q(i)} \right) P(i)
$$
>
>$\quad$ $\bullet$ Continuous Random Variable인 경우
$$
D_{KL}(P \| Q) = \int_i log_2 \left( \frac{p(x)}{q(x)} \right) p(x) dx
$$

#### (6.2) KL-divergence의 특성

$\quad$ **특성 1**: KLD는 Non-negative임
$$
D_{KL}(P \| Q) \ge 0
$$

[증명]<br>
다음 그림과 같은 convex function $f(x)$를 생각함

convec function의 정의에 따라 $alpha \ge 0$에 대해 다음식이 성립함
$$
f(\alpha x_1 + (1 - \alpha) x_2) \le \alpha f(x_1) + (1-\alpha) f(x_2)
$$
$\qquad \qquad \Longrightarrow$
$$
f(E(X)) \le E[f(X)]
$$

$\qquad \qquad \qquad \qquad $ $\because$ $(\alpha) + (1 -  \alpha)=1$이고 $\alpha \ge 0$
<br><br>

$$
\begin{align}
D_{KL}(P \| Q) =& \int log_2 \left( \frac{p(x)}{q(x)} \right) p(x) dx  \\
               =& \int log_2 \left( - \frac{q(x)}{p(x)} \right) p(x) dx \\
               \ge & log_2 \int \left( - \frac{q(x)}{p(x)} \right) p(x) dx \\
               =& - log_2 (1) \\
               =& 0
\end{align}
$$

따라서 
$$
D_{KL}(P \| Q)\ge 0
$$

$\quad$ **특성 2**: KLD는 거리 개념이 아니고 asymmetric한 measure임
$$
D_{KL}(P \| Q) \neq D_{KL}(Q \| P)
$$

[증명]<br>
$D_{KL}(P \| Q) - D_{KL}(Q \| P)$를 계산해보자<br>

$$
\begin{align}
D_{KL}(P \| Q) - D_{KL}(Q \| P) =& H(p, q) -H(p) - \left[ H(q, p) -H(q) \right] \\
                                =& H(q) - H(p)  (\qquad \because H(p, q) = H(q, p))) \\
                                \neq & 0 
\end{align}
$$

>
>$$
D_{KL} (P \| Q) \simeq \frac{1}{n} \sum_{k=1}^{n} \left( log p(y_i) - log q(y_n) | \theta) \right)
$$
>
>(iv) $log p(x_n)$은 $\theta$에 대해 독립이고, $− log q(x_n|\theta)$ 는 training set으로 얻은 $q(Y|\theta)$ 분포 하에서 $\theta$ 에 대한 negative log likelihood임
>
>(v) 그러므로 KL-divergence를 minimize하는 것이 likelihood를 maximize하는 것과 같다는 것을 알 수 있음
>
>$\blacksquare$ Variational Auto Encoder에서 Latent vector $z^{(c)}$의 분포 $p \left( z^{(c)} | z^{(1)} \right)$를 $q \left( z^{(c)} \right)$로 근사화하는 방법을 사용함

#### (6.4) Jenson-Shannon Divergence
- KL-Divergence는 근사된 확률 분포가 얼마나 원본과 비슷한지를 측정하는 척도임<br>

- 더 엄밀하게 정의하면 
$D_{KL}( P\| Q)$는 확률분포 $P$를 확률분포 $Q$로 근사화할 때, 정보량의 차이(정보손실량)의 기댓값임<br><br>

- 그리고 KL-Divergence는 Symmetric $ D_{KL}(P \| Q) \neq D_{KL}(Q \| P) $하지 않음
 
- Jenson-Shannon Divergence는 두 분포간의 차이를 계산할 때 symmetric하게 사용할 수 있도록 변형한 measure로 거리의 개념을 가짐
 <br><br>
 
$$
JSD(P, Q) = \frac{1}{2} D_{KL} \left( P  \left\| \frac{P+Q}{2} \right. \right) 
           +
            \frac{1}{2} D_{KL} \left( Q  \left\| \frac{P+Q}{2} \right. \right) 
$$

#### (7) Cross-Entropy

#### (7.1) Cross-Entropy의 정의

- 인공지능에서 다루는 예측 모형은 참인 실제 분포인 $p(y)$를 모르고 $p(y)$로 부터 sampling된 data $y_1, ..., y_n$을 이용하여 $p(y)$ 분포를 예측하고자 함

- 예측 모델링을 통해 구한 분포를 $q(y)$ 라고 할 때, 참인 실제 분포인 $p(y)$를 추정한  $q(y)$ 분포를 만들었을 때, 추정한 $q(y)$가 $p(y)$에 얼마나 잘 근사화 되었는지 판단하기 위해 다음과 같이 정의된 cross-entropy를 사용함 

$$ 
\begin{align}
H_p(q) =& E_p \left( log \frac{1}{q} \right)  \\\\
       =& 
\begin{cases} 
\sum_i log \frac{1}{q(i)} p(i)     &       if \ i가 \ 이산형확률변수일 \ 때  \\\\
\int_x log \frac{1}{q(y)} p(y)dy   &       if \ i가 \ 연속형확률변수일 \ 때
\end{cases}
\end{align}
$$ 

#### (7.2) Cross-Entropy와 KLD의 관계

- Kullback-Leibler Divergence $D_{KL}(P \| Q)$는 Cross-Entropy $H_p(q)$에 상수를 더한 값이므로 두 measure는 같은 measure로 사용할 수 있음

$$
\begin{align}
D_{KL}(P \| Q) =& E_p \left( log \frac{p(Y)}{q(Y)} \right) \\
               =& E_p \left( log \frac{1}{q(Y)} \right)
                +
                  E_p \left( log {p(Y)} \right) \\
               =& H_p(q) - H(p)
\end{align}
$$

$\qquad \qquad \Longrightarrow$

$$
H_p(q) = D_{KL}(P \| Q) + H(p)
$$

## 4. 왜 차원을 변화시키려는 것일까?

- Data 분석에서 data의 차원을 증가 또는 축소시키면서 얻는 장단점이 있음

## 5. 차원증가

### 5.1 차원 증가에 따른 장점

- 다음과 같은 예를 생각해 보자

$\qquad \qquad \qquad$ $
\begin{align}
y =& 어린이의 \ 키 \\
z =& 어린이의 \ 키에 \ 영향을 미치는 변수(feature)
\end{align}
$<br><br>

$\qquad$ 어린이의 키에 영향을 미치는 변수로 <br><br>
$\qquad \qquad \qquad z_1$  =  유전자<br>
$\qquad \qquad \qquad z_2$  =  1일  영양섭취량 <br><br>
$\qquad \quad$ 를 선택한 경우를 생각해 보자.

- 이 때 어린이의 키 $y$를 설명하는 변수로 $z_1$만 선택한 (Model 1)과 $z_1$과 $z_2$ 2개 변수를 선택한 (Model 2)를 각각 생각해보면 다음과 같음(선형 Model을 가정한 경우)

$$
\begin{align}
Model \ 1: y =& f_1(z_1)  \\
             =& w_0 + w_1 z_1 + \varepsilon \\\\
Model \ 2: y =& f_2(z_1, z_2) \\
             =& w_0 + w_1 z_1 + w_2 z_2 + \varepsilon
\end{align}
$$

$\qquad$ Model 2가 어린이의 키를 좀 더 정교하게 설명할 수 있을 것임. 즉 어린이의 키 $y$ 값의 변동 $(Var(y))$의 이유를 더 정교하게 설명할 수 있음

- 이와 같이 차원의 증가로 관심 측정치 $y$를 좀 더 정확하게 설명할 수 있는 가능성이 커짐 

### 5.2 차원 증가에 따른 단점

#### (1) 차원 증가에 따른 data 간의 거리 증가
- data가 $n$개 있을 때 차원이 증가하게 되면 data간의 간격은 멀어지게 됨
- 즉, data 간의 거리가 멀어지는 현상을 data가 sparce해진다고 말함
- $n$개의 data가 $k$ 차원에 존재할 때,  data간의 평균거리를 $d_k$라고 표현하면 $k$ 차원 내에서 data간의 평균거리는 다음과 같이 표현할 수 있음

 설명 | 점 | 선 | 면 | 체적 | Hyper Dimension
:----:|:----:|:----:|:----:|:----:|:----:
차원 |  0  | 1 | 2 | 3 | $$n$$
각 점간의 거리 |  |   |  |   |   |  
각 점간의 거리 2 | $$ \qquad \quad d_0 = 0 \qquad \quad $$ | $$ \qquad \quad d_1>0 \qquad \quad $$ | $$d_2 >> d_1$$ | $$d_3 >>> d_2 >> d_1$$ | $$d_n >>>> d_3 >>> d_2 >> d_1$$

#### (2) Locallity가정의 붕괴

- 통계적학습에서 뿐 아니라 data를 통해 어떤 사실 또는 규칙을 알아내고자 하는 학문에서는 많은 경우, <span style="color:red">"인접한 data들 간에 유사한 성질을 가진다"</span> 라는 성질을 이용하게 됨 <br>
\- 대표적으로 image 분석에서는 'neighborhood' 개념을 적용함<br>
\- Convolution 연산은 'neighborhood' 개념을 사용한 연산이라고 할 수 있음 <br>
\- 다른 예로 $kNN(k-nearest neighbor)$은  <span style="color:red">"인접한 data들 간에 유사한 성질을 가진다"</span> 라는 가정을 <br> $ \ \ $ 전제로 개발된 algorithm임<br>
\- $kNN$ 뿐 아니라 많은 algorithm에서 "인접한 data들 간에 유사한 성질을 가진다"라고 가정함<br><br>

- 하지만 차원이 높아지면 data 간의 거리가 멀어짐에 따라 <span style="color:red">"인접한 data들 간에 유사한 성질을 가진다"</span> 라는 가정은 더 이상 성립하지 않게 됨

#### kNN(k-Nearest Neighbor)의 Classification

- 즉, 한 점 $ \left( x_1^*, x_2^* \right)$의 일정 반경 $\epsilon_1$ 내에 $k$개의 다른 data들이 없을 수 있기 때문에 $k$개의 neighborhood point를 <br> $ \ \ $ 확보하기 위해 $\epsilon_1$을 $ d (\gg \epsilon_1)$로 변경해야함<br><br>


- 이렇게 매우 큰 $ d $로 확대한 반경  내에 포함된 $k$개 점들을 확보했을 때,  반경 $d$ 내의 임의의 점 $ \left( x_1, x_2 \right)$가 $ \left( x_1^*, x_2^* \right)$와 같은 성질을 가진다고 말할 수 있을까?

#### (3) 더 많은 Data 처리 자원 필요

- data의 type이 숫자에서 image(또는 영상), 소리(또는 음향)등 크기가 큰 data로 빠르게 확대되고 있음

- 기존 Feature data $Z = \left( z_1, z_2, ..., z_k \right)$에 새로운 feature $z_a$를 추가하는 경우, 다음과 같은 4가지 경우가 존재함

번호 | 기존 $z_i$와의 <br> 함수관계 | $y$와의 <br> 함수관계 | 설명
:----:|:----:|:----:|:----
<br> (1) | <br> $\times$ | <br> $\times$| <br> (1) 무의미한 feature도입임 <br> (2) $y$에 대한 설명력 증대나 $z_i$들간의 관계성 설명력 증가없이 계산 복잡도만 증가 
<br> (2) | <br> $\times$ | <br> $\bigcirc$ | <br> (1) 새로운 feature $z_a$의 도입으로 차원이 높아지고 계산량이 늘어나지만,  <br> (2) Label $y$의 변동  $Var(y)$을  추가로 설명할 수 있는 feature $z_a$가 도입되므로 <br> $\quad$ feature $z_a$의 도입효과를 볼 수 있음
<br> (3) | <br> $\bigcirc$ | <br> $\times$ | <br> (1) 기존 $Z$와의 상관성이 크지 않다면 $z_a$ 도입을 통해 $y$에 대한 설명력을 높일 수 잇음  <br> (2) 하지만 기존 $Z$와의 상관성이 크다면 $y$에 대한 설명력 증가는 없이 계산복잡도만 증가함
<br> (4) | <br> $\bigcirc$ | <br> $\bigcirc$ |  <br> (1) $z_a$와 $z_i$사이에 다중공선성(Multicollinearity)이 발생함 <br> (2) 다중공선성(Multicollinearity)이 발생하면 다음과 같은 문제가 발생함 <br> $\quad$ (i) 추가된 특정 feature  $z_a$와 관련한 parameter $\theta_a$의 추정치 $ \hat \theta_a$가 상관성을 가지는 다른 feature  $z_i$와 <br>  $\qquad$ 관련한 parameter $\theta_i$의 추정치 $\hat \theta_i$에 크게 의존하게 됨 <br> $\quad$ (ii) 상관성을 가지는 feature와 관련한  parameter $ \theta $의 추정치 $ \hat \theta $의 variance $Var \left( \hat \theta \right)$가 매우 커짐 <br> $\quad$ (iii) 상관성을 가지는 feature를 도입한 경우, 새로운 feature $z_a$를 도입하였음에도 Label $y$의 변동을  <br> $\qquad$ 추가로 설명할 수 있는 양이 매우 작음 <br> $\quad$ (vi) 기존 feature $z_i$와 상관성이 높은 feature $z_a$가 model에 이미 포함되어 있는지 여부에 따라, 관심 <br>  $\qquad$ feature  $z_a$와  parameter 관련한 $   \theta_a $의 추정값이 크게 달라짐 

#### [참고] 상관관계 

- 통계학에서 2개 feature들 $z_i$와 $z_j$ 간의 선형관계를 파악하는 방법으로 상관계수가 있음
- 상관계수는 다음과 같이 정의 됨

$$
\rho_{z_i, z_j} = \frac{Cov \left( z_i, z_j \right) }{\sigma_{z_i} \times \sigma_{z_j}}
$$

$\qquad$ 여기서 $\sigma_{z_i}$와 $\sigma_{z_j}$는 $z_i$와 $z_j$의 표준편차임 <br><br>

$\qquad - \ $ $Cov \left( z_i, z_j \right)$는 $z_i$와 $z_j$의 선형관계성을 평가하는 척도임에 주의 <br>
$\qquad \quad $ (1) 즉, 부적절한 차원증가로 발생하는 문제점 중 상관성이 큰 feature의 도입으로 발생하는 문제는 선형 <br> $\qquad \qquad $ 관계성에 대한 논의에 국한 된다는 점에 유의해야하고, <br>
$\qquad \quad $ (2) $z_a$와 $z_i$간의 비선형 관계에 대해서는 성에 대해서는  $Cov \left( z_i, z_j \right) \approx 0$이라는 의미가 관계성이 없음을   <br> $\qquad \qquad $ 의미하는 것은 아님. 다만 선형관계가 없다는 의미이고 비선형 관계성은 존재할 수 있음에 주의

## 6. 차원축소 방법

- 기존 feature $z_i$들 중 무의미한 feature $z_l$이 존재한다면 불필요한 계산복잡도가 유발된 상황이므로 불필요한 $z_l$을 제거하여 차원을 축소해야함
